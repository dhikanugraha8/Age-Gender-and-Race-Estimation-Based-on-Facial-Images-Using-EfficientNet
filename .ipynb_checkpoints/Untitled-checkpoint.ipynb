```python
import os
import re
from collections import Counter

# Path to the dataset
dataset_path = 'dataset'
pattern = r"(\d+)_(\d)_(\d)_(\d+).jpg"

# Get all image files
all_files = [f for f in os.listdir(dataset_path) if f.endswith('.jpg')]

# Extract ages from filenames
ages = []
for file in all_files:
    match = re.match(pattern, file)
    if match:
        age = int(match.group(1))
        ages.append(age)

# Count occurrences of each age
age_counts = Counter(ages)

# Display the counts
for age, count in sorted(age_counts.items()):
    print(f"Age {age}: {count} images")
```

```python
import os
import re
import shutil
from collections import defaultdict
import random

# Path to the original dataset and the new downsampled dataset
original_dataset_path = 'dataset_full'
downsampled_dataset_path = 'dataset'

# Pattern to match the age in filenames
pattern = r"(\d+)_(\d)_(\d)_(\d+).jpg"

# Create the downsampled dataset folder if it doesn't exist
os.makedirs(downsampled_dataset_path, exist_ok=True)

# Dictionary to store file paths grouped by age
age_dict = defaultdict(list)

# Get all image files
all_files = [f for f in os.listdir(original_dataset_path) if f.endswith('.jpg')]

# Group files by age
for file in all_files:
    match = re.match(pattern, file)
    if match:
        age = int(match.group(1))
        age_dict[age].append(file)

# Downsample and copy files to the new folder
for age, files in age_dict.items():
    # If there are more than 100 images for an age, randomly select 100
    if len(files) > 100:
        files = random.sample(files, 100)
    # Copy the selected files to the new dataset folder
    for file in files:
        src = os.path.join(original_dataset_path, file)
        dst = os.path.join(downsampled_dataset_path, file)
        shutil.copy(src, dst)

print("Downsampling complete.")
```

```python
import tensorflow as tf
import os
import re
import random
from tensorflow.keras import layers

# Path to the dataset
dataset_path = 'dataset'
image_size = (224, 224)
batch_size = 32

# Regular expression to parse the filename
pattern = r"(\d+)_(\d)_(\d)_(\d+).jpg"

# Function to parse the label from the filename
def parse_label(filename):
    match = re.match(pattern, filename)
    if match:
        age = int(match.group(1))
        return age
    else:
        raise ValueError(f"Filename {filename} does not match the format")

# Function to load image and label
def load_image(filename, label):
    image = tf.io.read_file(filename)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, image_size)
    image = tf.cast(image, tf.float32) / 255.0
    return image, label

# Function to create a dataset from image files
def create_dataset(image_files):
    labels = [parse_label(os.path.basename(file)) for file in image_files]
    ages = tf.constant(labels, dtype=tf.int32)

    dataset = tf.data.Dataset.from_tensor_slices((image_files, ages))
    dataset = dataset.map(lambda x, y: (load_image(x, y)), num_parallel_calls=tf.data.AUTOTUNE)
    dataset = dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return dataset
```

```python
# Get all image files
all_files = [os.path.join(dataset_path, f) for f in os.listdir(dataset_path) if f.endswith('.jpg')]
total_files = len(all_files)

# Shuffle files with a consistent seed
seed = 42
random.seed(seed)
random.shuffle(all_files)

# Split the data into train, validation, and test
train_split = 0.7
val_split = 0.15
test_split = 0.15

train_count = int(total_files * train_split)
val_count = int(total_files * val_split)

train_files = all_files[:train_count]
val_files = all_files[train_count:train_count + val_count]
test_files = all_files[train_count + val_count:]

# Create dataset for each subset
train_ds = create_dataset(train_files)
val_ds = create_dataset(val_files)
test_ds = create_dataset(test_files)
```

```python
import matplotlib.pyplot as plt

# Function to show image samples and labels
def show_image_samples(dataset, subset_name, num_samples=5):
    plt.figure(figsize=(15, 6))  # Increase image size
    for images, labels in dataset.take(1):  # Take the first batch from the dataset
        for j in range(num_samples):
            ax = plt.subplot(1, num_samples, j + 1)  # 1 row, num_samples columns
            plt.imshow(images[j].numpy())  # Display the image
            plt.title(f"Age: {labels[j].numpy()}", fontsize=100/num_samples)  # Display the label
            plt.axis("off")  # Hide the axes
    
    plt.suptitle(f"{subset_name} Samples", fontsize=20)  # Title for each subset
    plt.show()
    print('\n')

# Display some image samples and labels from each subset
show_image_samples(train_ds, "Training",5)
show_image_samples(val_ds, "Validation")
show_image_samples(test_ds, "Testing")
```

```python
# Image augmentation with flip and rotation
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
])
```

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras import Model, layers
import tensorflow as tf

# Load ResNet50 model without the top layer
base_model = ResNet50(input_shape=(224, 224, 3), include_top=False, weights='imagenet')
```

```python
base_model.summary()
```

```python
# Freeze the base model layers except for the last 10 for fine-tuning
for layer in base_model.layers:
    if 'conv5_block3' not in layer.name:
        layer.trainable = False

# Prepare input and include augmentation in the architecture
inputs = tf.keras.Input(shape=(224, 224, 3))
x = data_augmentation(inputs)
x = base_model(x, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(128, activation='relu')(x)
x = layers.Dropout(0.2)(x)

# Output for age (regression)
age_output = layers.Dense(1, name='age')(x)

# Model with a single output for age
model = Model(inputs=inputs, outputs=age_output)
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# Prepare EarlyStopping callback
early_stopping = EarlyStopping(
    monitor='val_loss', 
    patience=5,          
    restore_best_weights=True 
)

# Compile the model
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    loss='mse',
    metrics=['mae']
)
```

```python
# Fine-tune with fit
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=30,
    callbacks=[early_stopping]
)
```
