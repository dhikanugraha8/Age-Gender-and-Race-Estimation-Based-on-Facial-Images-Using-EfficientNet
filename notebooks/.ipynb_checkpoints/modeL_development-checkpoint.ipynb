```python
import os
import re
import random
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

# Regular expression to parse the filename
dataset_path ='../data'
pattern = r"(\d+)_(\d)_(\d)_(\d+).jpg"

# Fungsi untuk mengurai label dari nama file
def parse_label(filename):
    match = re.match(pattern, filename)
    if match:
        age = int(match.group(1))
        gender = int(match.group(2))
        race = int(match.group(3))
        return age, gender, race
    else:
        raise ValueError(f"Filename {filename} tidak sesuai format")

from collections import Counter
import matplotlib.pyplot as plt

# Fungsi untuk plot distribusi data berdasarkan label
def plot_initial_distribution(files):
    ages = [parse_label(os.path.basename(f))[0] for f in files]  # Umur
    genders = [parse_label(os.path.basename(f))[1] for f in files]  # Gender
    races = [parse_label(os.path.basename(f))[2] for f in files]  # Race
    
    # Plot distribusi race
    plt.figure(figsize=(8, 5))
    race_counts = Counter(races)
    plt.bar(race_counts.keys(), race_counts.values(), color=['gray', 'black', 'yellow', 'red', 'green'])
    plt.xticks([0, 1, 2, 3, 4], ['White', 'Black', 'Asian', 'Indian', 'Others'])
    plt.title('Race Distribution in Dataset (Before Downsampling)')
    plt.xlabel('Race')
    plt.ylabel('Count')
    plt.show()

    # Plot distribusi age
    plt.figure(figsize=(8, 5))
    plt.hist(ages, bins=20, color='blue', edgecolor='black')
    plt.title('Age Distribution in Dataset (Before Downsampling)')
    plt.xlabel('Age')
    plt.ylabel('Frequency')
    plt.show()

    # Plot distribusi gender
    plt.figure(figsize=(8, 5))
    gender_counts = Counter(genders)
    plt.bar(gender_counts.keys(), gender_counts.values(), color=['blue', 'pink'])
    plt.xticks([0, 1], ['Male', 'Female'])
    plt.title('Gender Distribution in Dataset (Before Downsampling)')
    plt.xlabel('Gender')
    plt.ylabel('Count')
    plt.show()

# Get all image files
all_files = [os.path.join(dataset_path, f) for f in os.listdir(dataset_path) if f.endswith('.jpg')]

# Panggil fungsi untuk menampilkan distribusi awal
plot_initial_distribution(all_files)
```

```python
import numpy as np
import pandas as pd

# Fungsi untuk menghitung distribusi usia dalam rentang yang diinginkan
def age_distribution(files, bin_edges):
    ages = [parse_label(os.path.basename(f))[0] for f in files]  # Ekstrak usia dari filename

    # Menggunakan numpy untuk menghitung histogram dengan rentang yang diberikan
    hist, bins = np.histogram(ages, bins=bin_edges)

    # Menampilkan hasil distribusi usia dalam bentuk tabel
    age_ranges = [f'{int(bins[i])}-{int(bins[i+1])}' for i in range(len(bins)-1)]
    distribution = pd.DataFrame({'Age Range': age_ranges, 'Count': hist})
    
    print(distribution)

    # Plot distribusi usia dalam bentuk barchart
    plt.figure(figsize=(10, 6))
    plt.bar(age_ranges, hist, color='blue', edgecolor='black')
    plt.xticks(rotation=45)
    plt.title('Age Distribution in Dataset (With Ranges)')
    plt.xlabel('Age Range')
    plt.ylabel('Count')
    plt.show()

# Rentang usia yang diinginkan, misal setiap 5 tahun
age_bins = [0, 5, 10, 15, 20, 25, 30, 35, 40, 50, 60, 70, 80, 90, 100, 120]

# Panggil fungsi untuk menghitung dan menampilkan distribusi usia
age_distribution(all_files, age_bins)
```

```python
# Fungsi untuk melakukan downsampling dengan mempertimbangkan race dan age
def downsample_by_race_and_age(files, target_per_class=2000):
    age_25_30_range = (25, 30)
    
    # Group data by race
    race_to_files = {i: [] for i in range(5)}
    for f in files:
        _, _, race = parse_label(os.path.basename(f))
        race_to_files[race].append(f)

    # New list to store selected files after downsampling
    selected_files = []

    for race, race_files in race_to_files.items():
        if race == 4:  # Skip 'Others' race, do not downsample it
            selected_files.extend(race_files)
            continue

        if len(race_files) > target_per_class:
            # Pisahkan file dengan rentang umur 25-30
            age_25_30_files = [f for f in race_files if age_25_30_range[0] <= parse_label(os.path.basename(f))[0] <= age_25_30_range[1]]
            other_age_files = [f for f in race_files if f not in age_25_30_files]
            
            # Prioritaskan pengurangan dari rentang usia 25-30
            if len(age_25_30_files) > 2000:
                downsampled_25_30_files = random.sample(age_25_30_files, 2000)  # Kurangi hingga 2000 gambar
                selected_files.extend(downsampled_25_30_files)
            else:
                # Masukkan semua file usia 25-30, jika < 2000 gambar
                selected_files.extend(age_25_30_files)
            
            # Jika masih perlu lebih banyak pengurangan, ambil dari usia lainnya
            remaining_needed = target_per_class - len(age_25_30_files)
            if remaining_needed > 0:
                downsampled_other_files = random.sample(other_age_files, remaining_needed)
                selected_files.extend(downsampled_other_files)
        else:
            # Jika jumlah file sudah kurang dari target, tidak perlu downsampling
            selected_files.extend(race_files)

    return selected_files

# Lakukan downsampling
downsampled_files = downsample_by_race_and_age(all_files, target_per_class=2000)
```

```python
import tensorflow as tf
import os
import re
import random
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras import Input, Model, layers
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping

# Path to the dataset
dataset_path = '../data_downsampled'
image_size = (200, 200)
batch_size = 32

# Regular expression to parse the filename
pattern = r"(\d+)_(\d)_(\d)_(\d+).jpg"

# Function to parse the label from the filename
def parse_label(filename):
    match = re.match(pattern, filename)
    if match:
        age = int(match.group(1))
        gender = int(match.group(2))
        race = int(match.group(3))
        return age, gender, race
    else:
        raise ValueError(f"Filename {filename} does not match the format")

# Function to load image and label
def load_image(filename, label):
    image = tf.io.read_file(filename)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, image_size)
    image = tf.cast(image, tf.float32) / 255.0
    
    age, gender, race = label
    return image, (age, gender, race)

# Function to create a dataset from image files
def create_dataset(image_files):
    labels = [parse_label(os.path.basename(file)) for file in image_files]
    ages = tf.constant([label[0] for label in labels], dtype=tf.int32)
    genders = tf.constant([label[1] for label in labels], dtype=tf.int32)
    races = tf.constant([label[2] for label in labels], dtype=tf.int32)

    dataset = tf.data.Dataset.from_tensor_slices((image_files, (ages, genders, races)))
    dataset = dataset.map(lambda x, y: (load_image(x, y)), num_parallel_calls=tf.data.AUTOTUNE)
    dataset = dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return dataset
```

```python
# Get all image files
all_files = [os.path.join(dataset_path, f) for f in os.listdir(dataset_path) if f.endswith('.jpg')]
total_files = len(all_files)

# Shuffle files with a consistent seed
seed = 42
random.seed(seed)
random.shuffle(all_files)

# Split the data into train, validation, and test
train_split = 0.5
val_split = 0.25
test_split = 0.25

train_count = int(total_files * train_split)
val_count = int(total_files * val_split)

train_files = all_files[:train_count]
val_files = all_files[train_count:train_count + val_count]
test_files = all_files[train_count + val_count:]

# Create dataset for each subset
train_ds = create_dataset(train_files)
val_ds = create_dataset(val_files)
test_ds = create_dataset(test_files)
```

```python
# Function to show image samples and labels for age, gender, and race
def show_image_samples(dataset, subset_name, num_samples=5):
    plt.figure(figsize=(20, 6))
    gender_map = {0: 'Male', 1: 'Female'}
    race_map = {0: 'White', 1: 'Black', 2: 'Asian', 3: 'Indian', 4: 'Others'}
    
    for images, (ages, genders, races) in dataset.take(1):
        for j in range(num_samples):
            ax = plt.subplot(1, num_samples, j + 1)
            plt.imshow(images[j].numpy())
            age = ages[j].numpy()
            gender = genders[j].numpy()
            race = races[j].numpy()
            plt.title(f"Age: {age}\n Gender: {gender_map[gender]} \n Race: {race_map[race]}", fontsize=100/num_samples)
            plt.axis("off")
    
    plt.suptitle(f"{subset_name} Samples", fontsize=20)
    plt.show()
    print('\n')

# Display some image samples and labels from each subset
show_image_samples(train_ds, "Training",)
show_image_samples(val_ds, "Validation")
show_image_samples(test_ds, "Testing")
```

```python
# Function to plot the age distribution
def plot_age_distribution(dataset, subset_name):
    ages = []
    for _, (age_labels, _, _) in dataset:
        ages.extend(age_labels.numpy())
    
    plt.figure(figsize=(10, 6))
    plt.hist(ages, bins=range(0, 120, 5), edgecolor='black')
    plt.title(f"Age Distribution in {subset_name} Set")
    plt.xlabel("Age")
    plt.ylabel("Frequency")
    plt.show()

# Plot age distributions for training, validation, and test sets
plot_age_distribution(train_ds, "Training")
plot_age_distribution(val_ds, "Validation")
plot_age_distribution(test_ds, "Testing")
```

```python
# Function to plot the gender distribution
def plot_gender_distribution(dataset, subset_name):
    gender_count = {0: 0, 1: 0}
    for _, (_, gender_labels, _) in dataset:
        unique, counts = np.unique(gender_labels.numpy(), return_counts=True)
        for u, c in zip(unique, counts):
            gender_count[u] += c
    
    plt.figure(figsize=(6, 4))
    plt.bar(['Male', 'Female'], [gender_count[0], gender_count[1]], color=['blue', 'pink'])
    plt.title(f"Gender Distribution in {subset_name} Set")
    plt.xlabel("Gender")
    plt.ylabel("Count")
    plt.show()

# Plot gender distributions for training, validation, and test sets
plot_gender_distribution(train_ds, "Training")
plot_gender_distribution(val_ds, "Validation")
plot_gender_distribution(test_ds, "Testing")
```

```python
# Function to plot the race distribution
def plot_race_distribution(dataset, subset_name):
    race_count = {0: 0, 1: 0, 2: 0, 3: 0, 4: 0}
    for _, (_, _, race_labels) in dataset:
        unique, counts = np.unique(race_labels.numpy(), return_counts=True)
        for u, c in zip(unique, counts):
            race_count[u] += c
    
    plt.figure(figsize=(8, 6))
    plt.bar(['White', 'Black', 'Asian', 'Indian', 'Others'], 
            [race_count[i] for i in range(5)], color=['gray', 'black', 'yellow', 'brown', 'green'])
    plt.title(f"Race Distribution in {subset_name} Set")
    plt.xlabel("Race")
    plt.ylabel("Count")
    plt.show()

# Plot race distributions for training, validation, and test sets
plot_race_distribution(train_ds, "Training")
plot_race_distribution(val_ds, "Validation")
plot_race_distribution(test_ds, "Testing")
```

```python
# Data augmentation layer
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomContrast(0.1)
])
```

```python
# Create Model
input_shape = (200, 200, 3)  # sesuai input shape
inputs = Input(input_shape)

# Convolutional layers
inputs = Input(shape=input_shape)
x = data_augmentation(inputs)  # Augmentasi di dalam arsitektur
x = layers.Conv2D(32, kernel_size=(3, 3), activation='relu')(x)
x = layers.MaxPooling2D(pool_size=(2, 2))(x)
x = layers.Conv2D(64, kernel_size=(3, 3), activation='relu')(x)
x = layers.MaxPooling2D(pool_size=(2, 2))(x)
x = layers.Conv2D(128, kernel_size=(3, 3), activation='relu')(x)
x = layers.MaxPooling2D(pool_size=(2, 2))(x)
x = layers.Conv2D(256, kernel_size=(3, 3), activation='relu')(x)
x = layers.MaxPooling2D(pool_size=(2, 2))(x)

x = layers.Flatten()(x)

# Dense layers for each output
dense_age = layers.Dense(256, activation='relu', kernel_regularizer=l2(0.01))(x)
dropout_age = layers.Dropout(0.3)(dense_age)
age = layers.Dense(1, activation='relu', name='age')(dropout_age)

dense_gender = layers.Dense(256, activation='relu', kernel_regularizer=l2(0.01))(x)
dropout_gender = layers.Dropout(0.3)(dense_gender)
gender = layers.Dense(1, activation='sigmoid', name='gender')(dropout_gender)

dense_race = layers.Dense(256, activation='relu', kernel_regularizer=l2(0.01))(x)
dropout_race = layers.Dropout(0.3)(dense_race)
race = layers.Dense(5, activation='softmax', name='race')(dropout_race)
    
# Model with two outputs
model = Model(inputs=inputs, outputs=[age,gender,race])
```

```python
model.compile(
    optimizer='adam',
    loss={'age': 'mse', 'gender': 'binary_crossentropy', 'race': 'sparse_categorical_crossentropy'},
    metrics={'age': 'mae', 'gender': 'accuracy', 'race': 'accuracy'}
)
```

```python
model.summary()
```

```python
# Early Stopping
early_stopping = EarlyStopping(
    monitor='val_loss', 
    patience=5,          
    restore_best_weights=True 
)

# Model fitting
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=30,
    callbacks=[early_stopping]
)
```

```python
import matplotlib.pyplot as plt
import random

# Ambil batch acak dari test_ds
for images, labels in test_ds.take(1):
    batch_size = images.shape[0]
    random_indices = random.sample(range(batch_size), 5)
    random_images = images.numpy()[random_indices]

    # Memisahkan age, gender, dan race dari labels
    ages = labels[0].numpy()[random_indices]
    genders = labels[1].numpy()[random_indices]
    races = labels[2].numpy()[random_indices]

    # Lakukan prediksi
    predictions = model.predict(random_images)
    pred_ages = predictions[0].flatten()
    pred_genders = predictions[1].flatten()
    pred_races = predictions[2].argmax(axis=1)
    pred_race_confidences = predictions[2].max(axis=1)

    # Plot gambar dan prediksi
    plt.figure(figsize=(15, 15))
    for i in range(5):
        true_age = ages[i]
        true_gender = 'Male' if genders[i] == 0 else 'Female'
        true_race = ['White', 'Black', 'Asian', 'Indian', 'Others'][races[i]]

        pred_age = int(pred_ages[i])
        pred_gender = 'Male' if pred_genders[i] < 0.5 else 'Female'
        gender_confidence = 1 - pred_genders[i] if pred_genders[i] < 0.5 else pred_genders[i]
        pred_race = ['White', 'Black', 'Asian', 'Indian', 'Others'][pred_races[i]]
        race_confidence = pred_race_confidences[i]

        plt.subplot(1, 5, i + 1)
        plt.imshow(random_images[i])
        plt.title(f"Age: {true_age}\nGender: {true_gender}\nRace: {true_race}")
        plt.axis('off')  # Menghilangkan penggaris pixel
        
        # Tambahkan label di bawah gambar menggunakan plt.text
        plt.text(0.5, -0.1, 
                 f"Age Pred: {pred_age}\nGender Pred: {pred_gender} ({gender_confidence:.2f})\nRace Pred: {pred_race} ({race_confidence:.2f})",
                 fontsize=12, ha='center', va='top', transform=plt.gca().transAxes)

    plt.show()
```
